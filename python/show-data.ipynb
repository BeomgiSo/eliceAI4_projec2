````python
import re
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')

def text_preprocessing(text):
    # 영어 전처리 
    only_english = re.sub('[^a-zA-Z]', ' ', text)
    
    only_english = only_english.replace('game', '')
    only_english = only_english.replace('like', '')
    only_english = only_english.replace('play', '')
    
    no_capitals = only_english.lower().split()

    stops = set(stopwords.words('english'))
    no_stops = [word for word in no_capitals if not word in stops]

    stemmer = nltk.stem.SnowballStemmer('english')
    stemmer_words = [stemmer.stem(word) for word in no_stops]

    # 공백으로 구분된 문자열로 결합하여 결과 반환
    return ' '.join(stemmer_words)

# processed_comment = text_preprocessing(text)
# print(text)
````

````python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
import googletrans

translator = googletrans.Translator()


def get_word_cloud(comments, id):
    ko_arr = {}
    print('---translate---')
    # for item in comments:
    #     print('*', end=' ')
    #     translated = translate_to_ko(item)
    #     ko_arr.append(translated)
        
    counts_word = Counter(comments)
    # sorted_counter = sorted(counts_word.items(), key=lambda x: x[1], reverse=True)
    tags = counts_word.most_common(40)
    for i in tags:
        print('.', end='')
        print(i[0])
        try:
            translated_word = translator.translate(str(i[0]), dest='ko')
            print(translated_word.text)
        except:
            continue
        # translated = translate_to_ko(i[0])
        print(translated_word.text)
        ko_arr[translated_word.text] = i[1]
    wordcloud = WordCloud(
        width=1200,
        height=800,
        background_color='white',
        font_path='./NanumGothic.ttf'
    )

    wordclouds = wordcloud.generate_from_frequencies(dict(ko_arr))
    wordclouds.to_file(f'./wordClouds/{id}.png')
    # plt.figure(figsize=(10, 8))
    # plt.axis('off')
    # plt.imshow(wordclouds)
    # plt.show()
````

````python
def make_word_cloud(id):
    board_game = dropna_bgg_15[dropna_bgg_15['ID']==id]
    comment_array = []
    print('---preprocessing---')
    for i in board_game.iterrows():
        preprocess_string = text_preprocessing(i[1]['comment']).strip().split()
        comment_array += preprocess_string
    get_word_cloud(comment_array, id)


# make_word_cloud(30549)
for ids in id_list_by_rank:
    make_word_cloud(int(ids))
    print('one fin')
````

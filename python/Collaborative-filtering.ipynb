```python
df=pd.DataFrame(new_df,columns=["user","ID","rating","name"])
df=df.dropna()
df["user"] = LabelEncoder().fit_transform(df["user"])
```

```python
game_user_rating=df.pivot_table('rating',index='ID',columns='user')
```

```python
game_user_rating.head()
```

```python
game_user_rating.fillna(0,inplace=True)
```

```python
similarity_rate = cosine_similarity(game_user_rating,game_user_rating)
print(similarity_rate)
```

```python
similarity_rate_df=df = pd.DataFrame(
data=similarity_rate,
index = game_user_rating.index,
columns=game_user_rating.index)
```

```python
similarity_rate_df.head()
```

```python
def recommand_game(id):
    return similarity_rate_df[id].sort_values(ascending=False)[1:6]
```

```python
a=recommand_game(475)
```
